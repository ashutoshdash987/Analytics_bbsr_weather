```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('bbsr_weather.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
tavg_2019_mean = df['tavg'][0:365].mean()
tavg_2020_mean = df['tavg'][365:731].mean()
tavg_2021_mean = df['tavg'][731:1096].mean()
tavg_2022_mean = df['tavg'][1096:1342]
```

```python
df['tavg'][0:365].fillna(tavg_2019_mean,inplace=True)
df['tmin'][0:365].fillna(tavg_2019_mean,inplace=True)
df['tmax'][0:365].fillna(tavg_2019_mean,inplace=True)
df['wspd'].fillna(2.5,inplace=True)

df['tavg'][365:731].fillna(tavg_2020_mean,inplace=True)
df['tmin'][365:731].fillna(tavg_2020_mean,inplace=True)
df['tmax'][365:731].fillna(tavg_2020_mean,inplace=True)
df['wspd'][365:731].fillna(2.5,inplace=True)


df['tavg'][731:1096].fillna(tavg_2021_mean,inplace=True)
df['tmin'][731:1096].fillna(tavg_2021_mean,inplace=True)
df['tmax'][731:1096].fillna(tavg_2021_mean,inplace=True)
df['wspd'][731:1096].fillna(2.5,inplace=True)


df['tavg'][1096:1342].fillna(tavg_2022_mean,inplace=True)
df['tmin'][1096:1342].fillna(tavg_2022_mean,inplace=True)
df['tmax'][1096:1342].fillna(tavg_2022_mean,inplace=True)
df['wspd'][1096:1342].fillna(2.5,inplace=True)
```

```python
df['Prcp'].fillna(0,inplace=True)
```

```python
df.head()
```

```python
df['Date']= pd.to_datetime(df['Date'])
df.head()
```

```python
df_2019 = df.loc[(df.Date >= '2019-01-01') & (df.Date < '2020-01-01')]
df_2020 = df.loc[(df.Date >= '2020-01-01') & (df.Date < '2021-01-01')]
df_2021 = df.loc[(df.Date >= '2021-01-01') & (df.Date < '2022-01-01')]
df_2022 = df.loc[(df.Date >= '2022-01-01') & (df.Date < '2023-01-01')]
```

```python
max_temp_2019 = max(df_2019.tmax)
min_temp_2019 = min(df_2019.tmin)
```

```python
max_temp_2020 = max(df_2020.tmax)
min_temp_2020 = min(df_2020.tmin)
```

```python
max_temp_2021 = max(df_2021.tmax)
min_temp_2021 = min(df_2021.tmin)
```

```python
max_temp_2022 = max(df_2022.tmax)
min_temp_2022 = min(df_2022.tmin)
```

```python
max_temps = [max_temp_2019,max_temp_2020,max_temp_2021,max_temp_2022]
year = ['2019','2020','2021','2022']
```

```python
plt.bar(year,max_temps,tick_label = year,color='green')
plt.title('Max Temperature Year Wise')

for i,value in zip(year,max_temps):
    plt.text(i,value-3,str(value),ha='center',color = 'white')

plt.show()
```

```python
min_temps = [min_temp_2019,min_temp_2020,min_temp_2021,min_temp_2022]
```

```python
plt.bar(year,min_temps,tick_label = year,color='green')
plt.title('Min temperature year wise')

for i,value in zip(year,min_temps):
    plt.text(i,value-1,str(value),ha='center',color = 'white')

plt.show()
```

```python
tavg_2019 = df['tavg'][0:365]
date_2019 = df['Date'][0:365]

tavg_2020 = df['tavg'][365:731]
date_2020 = df['Date'][365:731]

tavg_2021 = df['tavg'][731:1096]
date_2021 = df['Date'][731:1096]

tavg_2022 = df['tavg'][1096:1342]
date_2022 = df['Date'][1096:1342]
```

```python
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
plt.plot(date_2019,tavg_2019)
plt.title("2019")

plt.subplot(2,2,2)
plt.plot(date_2020,tavg_2020)
plt.title("2020")

plt.subplot(2,2,3)
plt.plot(date_2021,tavg_2021)
plt.title("2021")

plt.subplot(2,2,4)
plt.plot(date_2022,tavg_2022)
plt.title("2022")
plt.suptitle("year_wise average temperature")

plt.show()
```

```python
# Finding the hottest day among this 4 years

hottest_day = df.loc[df.tmax == max(df.tmax)]
print(hottest_day.Date)
```

**April 04 2019 is the hottest day from these 4 years**

```python
# Finding the coldest day among this 4 years

coldest_day = df.loc[df.tmin == min(df.tmin)]
print(coldest_day.Date)
```

**January 01 2019 is the codlest day from these 4 years**
